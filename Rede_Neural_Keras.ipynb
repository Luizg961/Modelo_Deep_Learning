```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from keras.models import Sequential
from keras.layers import Dense
from tensorflow.keras.utils import to_categorical
from sklearn import datasets
```

```python
base = datasets.load_iris()
base
```

```python
previsores = base.data
classe = base.target
classe
```

```python
previsores
```

```python
classe_dumy = to_categorical(classe)
classe_dumy
```

```python
x_treino, x_teste, y_treino, y_teste = train_test_split(previsores,
                                                 classe_dumy,
                                                 test_size = 0.3,
                                                 random_state = 0)
```

```python
modelo = Sequential()

modelo.add(Dense(units = 5, input_dim = 4))

modelo.add(Dense(units = 4))

modelo.add(Dense(units=3, activation="softmax"))
```

```python
modelo.summary()
```

```python
modelo.compile(optimizer = "adam", loss = "categorical_crossentropy",
              metrics = ["accuracy"])

modelo.fit(x_treino, y_treino, epochs = 100,
          validation_data = (x_teste, y_teste))

#    ACCURACY MALHORA SE VOCE COLOCAR MAIS EPOCHS
#    COLOQUEI APENAS 100 PARA FINS DIDATICOS PARA NAO FICAR POLUIDO
```

```python
previsoes = modelo.predict(x_teste)
previsoes
```

```python
#CONVEÇAO PARA TRUE E FALSE MELHOR ENTENDIMENTO

previsoes = (previsoes > 0.05)
previsoes
```

```python
y_teste_matrix = [np.argmax(t) for t in y_teste]
Y_previsoes_matrix = [np.argmax(t) for t in previsoes]
```

```python
confusao = confusion_matrix(y_teste_matrix, Y_previsoes_matrix)
confusao
```
